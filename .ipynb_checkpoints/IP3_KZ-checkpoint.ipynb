```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Here I wrote a function called insert_to_dataframe() which reads the file, converts the date column to a date time value, and inserts a new column with the company name.

```python
def insert_to_dataframe(filename):
    df = pd.read_csv(filename)
    df['Date'] = pd.to_datetime(df['Date'])
    df.insert(0, 'Company', filename.split('.')[0])
    return df
```

2. Now, I ran all four files through the function to save them to dataframes. 

```python
df1 = insert_to_dataframe('50Hertz.csv')
```

```python
df2 = insert_to_dataframe('Amprion.csv')
```

```python
df3 = insert_to_dataframe('TenneTTSO.csv')
```

```python
df4 = insert_to_dataframe('TransnetBW.csv')
```

3. Using the concat function, I merged the four dataframes into one.

```python
merged = pd.concat([df1, df2, df3, df4])
```

4. Here I grouped the means of the merged dataframe by both the company and the year. 

```python
grouped = merged.groupby(['Company', merged['Date'].dt.year]).mean()
```

```python
grouped
```

5. By extracting the columns for midnight and noon from the grouped dataframe, I created two series for the separate data. Then, I created horizontal bar harts for each series and saved them to separate files. 

```python
midnight = grouped['0:00:00']
noon = grouped['12:00:00']
```

```python
midnight
```

```python
noon
```

```python
midnight.sort_values().plot.barh()
plt.title('TW/hours by Company/Year @ Midnight')
plt.savefig('midnight_barh.png')
```

```python
noon.sort_values().plot.barh(color='red')
plt.title('TW/hours by Company/Year @ Noon')
plt.savefig('noon_barh.png')
```

6. Here I combined the two bar graphs into a single graph by creating a figure and putting each graph into a subplot. Then, I saved the graph to a file.

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1)
midnight.sort_values().plot.barh(color='silver', ax=ax1)
ax1.set_title('TW/hours by Company/Year @ Midnight')
ax2 = fig.add_subplot(2,1,2)
noon.sort_values().plot.barh(color='gray', ax=ax2)
ax2.set_title('TW/hours by Company/Year @ Noon')
plt.savefig('combined_bar_chart.png')
```

7. Now I created a function called daily_view() which accepts a dataframe, company name, date, and color and creates an area graph for that company on that date. I did this by creating a list for all of the data on that specific day for that company, and a list of the column names which includes all fo the times that the data was collected. These lists were used as the y and x values for the graph. 

```python
def daily_view(dataframe, company, date, color1):
    line = merged.loc[(merged['Company'] == company) & (merged['Date'] == date)]
    times = list(line.columns)
    values = line.values.tolist()
    plt.plot(times[2:98], values[0][2:98], color=color1)
    plt.title('Power Production: '+company+'/'+date)
    plt.xlabel('Time of Day')
    plt.ylabel('TW/h Generated')
    plt.fill_between(times[2:98], values[0][2:98], 0, color=color1)
    plt.xticks(['0:00:00', '5:00:00', '10:00:00', '15:00:00', '20:00:00'])
    plt.show()
```

```python
daily_view(merged, '50Hertz', '2019-08-24', 'purple')
```

BONUS- Here I created a unified graph that contains four area charts for each company on the same day by creating a figure. 

```python
fig = plt.figure(figsize=(14, 12))

ax1 = fig.add_subplot(2,2,1)
line1 = merged.loc[(merged['Company'] == '50Hertz') & (merged['Date'] == '2019-08-24')]
times1 = list(line1.columns)
values1 = line1.values.tolist()
ax1.plot(times1[2:98], values1[0][2:98], color='red')
ax1.set_title('Power Production: 50Hertz/2019-08-24')
ax1.set_xlabel('Time of Day')
ax1.set_ylabel('TW/h Generated')
ax1.fill_between(times1[2:98], values1[0][2:98], 0, color='red')
ax1.set_xticks(['0:00:00', '5:00:00', '10:00:00', '15:00:00', '20:00:00'])

ax2 = fig.add_subplot(2,2,2)
line2 = merged.loc[(merged['Company'] == 'Amprion') & (merged['Date'] == '2019-08-24')]
times2 = list(line2.columns)
values2 = line2.values.tolist()
ax2.plot(times2[2:98], values2[0][2:98], color='green')
ax2.set_title('Power Production: Amprion/2019-08-24')
ax2.set_xlabel('Time of Day')
ax2.set_ylabel('TW/h Generated')
ax2.fill_between(times2[2:98], values2[0][2:98], 0, color='green')
ax2.set_xticks(['0:00:00', '5:00:00', '10:00:00', '15:00:00', '20:00:00'])

ax3 = fig.add_subplot(2,2,3)
line3 = merged.loc[(merged['Company'] == 'TenneTTSO') & (merged['Date'] == '2019-08-24')]
times3 = list(line3.columns)
values3 = line3.values.tolist()
ax3.plot(times3[2:98], values3[0][2:98], color='blue')
ax3.set_title('Power Production: TenneTTSO/2019-08-24')
ax3.set_xlabel('Time of Day')
ax3.set_ylabel('TW/h Generated')
ax3.fill_between(times3[2:98], values3[0][2:98], 0, color='blue')
ax3.set_xticks(['0:00:00', '5:00:00', '10:00:00', '15:00:00', '20:00:00'])

ax4 = fig.add_subplot(2,2,4)
line4 = merged.loc[(merged['Company'] == 'TransnetBW') & (merged['Date'] == '2019-08-24')]
times4 = list(line4.columns)
values4 = line4.values.tolist()
ax4.plot(times4[2:98], values4[0][2:98], color='purple')
ax4.set_title('Power Production: TransnetBW/2019-08-24')
ax4.set_xlabel('Time of Day')
ax4.set_ylabel('TW/h Generated')
ax4.fill_between(times4[2:98], values4[0][2:98], 0, color='purple')
ax4.set_xticks(['0:00:00', '5:00:00', '10:00:00', '15:00:00', '20:00:00'])

plt.savefig('unified_area_graphs.png')
```
